### 3. Define a Convolutional Neural Network (CNN) Model

We will define a simple CNN model. The input image size is `[3, 225, 225]`, so we'll adjust the layers accordingly. The output layer will have 2 units for 'Fruits' and 'Vegetables'.

```python
from torch import nn
import torch

class TinyVGG(nn.Module):
    """
    Model architecture copying TinyVGG from:
    https://poloclub.github.io/cnn-explainer/
    """
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
        super().__init__()
        self.block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,
                         stride=2)
        )
        self.block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            # The output features from `block_2` need to be calculated.
            # For input size 225x225, after two MaxPool2d layers (stride=2, kernel_size=2):
            # 225 / 2 = 112.5 -> 112 (after first maxpool, integer division)
            # 112 / 2 = 56 (after second maxpool)
            # So, the feature map size is hidden_units * 56 * 56
            nn.Linear(in_features=hidden_units*56*56,
                      out_features=output_shape)
        )

    def forward(self, x: torch.Tensor):
        x = self.block_1(x)
        x = self.block_2(x)
        x = self.classifier(x)
        return x

# Instantiate the model
torch.manual_seed(42)
model_0 = TinyVGG(input_shape=3, # number of color channels (RGB) - determined from image.shape[0]
                  hidden_units=10, # number of filters
                  output_shape=len(new_class_names)).to(device)

# Print the model summary
print(model_0)
```

```python
!pip install torchinfo
import torch.nn.functional as F

class SEBlock(nn.Module):
    """
    Squeeze-and-Excitation Block for channel attention.
    """
    def __init__(self, channel, reduction=5): # Changed reduction from 16 to 5
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channel, channel // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channel // reduction, channel, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c) # Squeeze operation
        y = self.fc(y).view(b, c, 1, 1) # Excitation operation
        return x * y.expand_as(x) # Scale feature maps


class TinyVGGWithAttention(nn.Module):
    """
    TinyVGG model with Squeeze-and-Excitation (SE) attention blocks.
    """
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int, reduction=5): # Changed reduction from 16 to 5
        super().__init__()
        self.block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            SEBlock(channel=hidden_units, reduction=reduction), # Add SE Block
            nn.MaxPool2d(kernel_size=2,
                         stride=2)
        )
        self.block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, 3, padding=1),
            nn.ReLU(),
            SEBlock(channel=hidden_units, reduction=reduction), # Add SE Block
            nn.MaxPool2d(2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            # Output features from `block_2` (same calculation as original TinyVGG)
            nn.Linear(in_features=hidden_units*56*56,
                      out_features=output_shape)
        )

    def forward(self, x: torch.Tensor):
        x = self.block_1(x)
        x = self.block_2(x)
        x = self.classifier(x)
        return x

# Instantiate the model with attention
torch.manual_seed(42)
model_attention = TinyVGGWithAttention(input_shape=3,
                                     hidden_units=10,
                                     output_shape=len(new_class_names)).to(device)

print(model_attention)

# Summarize the new model
# Assuming BATCH_SIZE is defined and image input size is 3, 225, 225
import torchinfo
torchinfo.summary(model_attention, input_size=(BATCH_SIZE, 3, 225, 225))
```

### 5. Create Training and Testing Functions

```python
from tqdm.auto import tqdm

def train_step(model: nn.Module,
               dataloader: DataLoader,
               loss_fn: nn.Module,
               optimizer: optim.Optimizer,
               device: torch.device) -> tuple[float, float]:
    """Trains a PyTorch model for a single epoch."""
    model.train()
    train_loss, train_acc = 0, 0
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)
        y_pred = model(X)
        loss = loss_fn(y_pred, y)
        train_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        y_pred_class = torch.argmax(torch.softmax(y_pred, dim=1), dim=1)
        train_acc += (y_pred_class == y).sum().item()/len(y_pred)
    train_loss = train_loss / len(dataloader)
    train_acc = train_acc / len(dataloader)
    return train_loss, train_acc

def test_step(model: nn.Module,
              dataloader: DataLoader,
              loss_fn: nn.Module,
              device: torch.device) -> tuple[float, float]:
    """Tests a PyTorch model for a single epoch."""
    model.eval()
    test_loss, test_acc = 0, 0
    with torch.inference_mode():
        for batch, (X, y) in enumerate(dataloader):
            X, y = X.to(device), y.to(device)
            test_pred_logits = model(X)
            loss = loss_fn(test_pred_logits, y)
            test_loss += loss.item()
            test_pred_labels = test_pred_logits.argmax(dim=1)
            test_acc += ((test_pred_labels == y).sum().item()/len(test_pred_labels))
    test_loss = test_loss / len(dataloader)
    test_acc = test_acc / len(dataloader)
    return test_loss, test_acc

print("Training and testing functions defined.")
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image

# Define the specific transform needed for this function to ensure consistent sizing
# This is a defensive measure to ensure all images are 225x225 before stacking,
# in case the dataset's own transforms are not consistently applied here.
transform_for_stacking = transforms.Compose([
    transforms.Resize((225, 225)),
    transforms.ToTensor() # Ensure it's a tensor
])

# This function is a PyTorch-compatible way to collect all images (features)
# and their corresponding labels from a Subset dataset object.
# It does NOT perform feature extraction via a pre-trained model, but rather
# gathers the raw preprocessed image tensors and their labels.
def get_pytorch_raw_data_and_labels(dataset_subset):
  all_features = []
  all_labels = []
  for image_pil, label_int in dataset_subset:
    # Explicitly apply the transform to ensure consistent size before stacking
    # Note: image_pil might be a tensor already if ToTensor was applied previously.
    # If it's a tensor, convert to PIL for Resize, then back to tensor.
    if isinstance(image_pil, torch.Tensor):
        # Convert tensor to PIL Image if it's already a tensor
        # This assumes the tensor is C, H, W and needs to be scaled to 0-255.
        # A more robust solution might pass the `transform_pipeline` directly.
        # For now, let's assume if it's a tensor, its values are 0-1 and it's C,H,W.
        # Convert to H,W,C for PIL, then back.
        image_pil = transforms.ToPILImage()(image_pil)

    # Apply the resizing transform
    processed_image_tensor = transform_for_stacking(image_pil)

    all_features.append(processed_image_tensor)
    # 'label_int' is an integer, convert to PyTorch tensor
    all_labels.append(torch.tensor(label_int))

  # Stack all images into one large tensor and all labels into another
  if all_features: # Check if the dataset is not empty
      return torch.stack(all_features), torch.stack(all_labels)
  else:
      print(f"Warning: Attempted to get features and labels from an empty dataset subset.")
      return torch.tensor([]), torch.tensor([]) # Return empty tensors

# These variables (new_train_data, new_test_data) are defined in the consolidated setup cell (b3546205).
# Ensure that cell has been executed before running this one.
train_features, train_labels = get_pytorch_raw_data_and_labels(new_train_data)
# Note: 'validation_dataset' was not defined in the notebook's current state.
# If a validation set is required, the dataset splitting logic needs to be adjusted in a prior cell.
test_features, test_labels = get_pytorch_raw_data_and_labels(new_test_data)

print(f"Shape of train_features: {train_features.shape} (all training images stacked)")
print(f"Shape of train_labels: {train_labels.shape} (all training labels stacked)")
print(f"Shape of test_features: {test_features.shape} (all testing images stacked)")
print(f"Shape of test_labels: {test_labels.shape} (all testing labels stacked)")
```

```python
import torch
from torchvision import transforms
from PIL import Image # Corrected import to use PIL.Image

# Assuming device, class_names, and model_0 are defined in previous cells
# Use the same transform pipeline as used for the training data
transform_pipeline_for_prediction = transforms.Compose([
    transforms.Resize((225, 225)), # Resize to the input size of the model
    transforms.ToTensor()          # Convert to PyTorch Tensor
])

def predict_image(image_path_or_data, model, class_names, device):
    model.eval() # Set model to evaluation mode

    # Handle cases where image_path_or_data is a file path or PIL Image
    if isinstance(image_path_or_data, str):
        img = Image.open(image_path_or_data).convert('RGB')
    elif isinstance(image_path_or_data, Image.Image):
        img = image_path_or_data
    else:
        raise TypeError("Input must be a file path (str) or a PIL Image.")

    # Apply transformations
    img_tensor = transform_pipeline_for_prediction(img).unsqueeze(0) # Add batch dimension
    img_tensor = img_tensor.to(device) # Move to the correct device

    with torch.inference_mode():
        output = model(img_tensor)

    # Get prediction probability and class label
    probabilities = torch.softmax(output, dim=1)
    predicted_label = torch.argmax(probabilities, dim=1).item()
    predicted_class_name = class_names[predicted_label]

    print("************************************\n\n")
    print(f"...........It is a {predicted_class_name}!")
    print(f"...........Confidence: {probabilities[0, predicted_label].item():.4f}")
    print("\n\n************************************\n\n")
```
